### Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url_1 = "https://raw.githubusercontent.com/pavel1shatalov/EDP-Final-Project/main/z1.txt"
url_2 = "https://raw.githubusercontent.com/pavel1shatalov/EDP-Final-Project/main/z2.txt"
url_3 = "https://raw.githubusercontent.com/pavel1shatalov/EDP-Final-Project/main/z3.txt"

z1 = np.array(np.loadtxt(url_1))
z2 = np.array(np.loadtxt(url_2))
z3 = np.array(np.loadtxt(url_3))

D1 = z1[:,0]
b1 = z1[:,1]

D2 = z2[:,0]
b2 = z2[:,1]

D3 = z3[:,0]
b3 = z3[:,1]
```

```python
plt.figure(figsize=(10,10))
plt.polar(b1, D1)
plt.polar(b2, D2)
plt.polar(b3, D3)
plt.xlim(0, np.pi/2)
plt.title('Trajectory in polar coordinates')
plt.legend(['z1', 'z2', 'z3'], loc = (1, 0.55))
plt.show()
```
